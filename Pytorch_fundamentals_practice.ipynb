### Matrix multiplicaton (dot product)

There are two main rules for matrix multiplication to work:

1) Inner dimensions must match

(2, 3) * (2, 2): ❌ 3 != 2

(2, 3) * (3, 2): ✅ 3 == 3

2) The resulting matrix has the shape of the outer dimension:

(4, 3) * (6, 3): (4, 6)

## Reshaping, stacking, squeezing and unsqueezing tensors

Reshaping:- Reshapes an input tensor to a defined shape

View:- Returns the view of an input tensor of certain shape but keeps the same memory as the origin

Stacking:- Combine multiple tensors on-top of each other.

Squeezing:- Removes all '1' dimensions from a tensor.

Unsqueezing:- Adds a '1' dimension to a target tensor.

Permute:- Return a view of the input with dimensions permuted (swapped) in a certain way.


The main point of all of these is to manipulate the tensors in certain way.

## Creating some known data using the Linear Regression formula.

```python
import torch
from torch import nn    # nn contains all of PyTorch's building blocks for Neural Networks
import matplotlib.pyplot as plt
from numpy import cos

# Check PyTorch version
torch.__version__
```

```python
# LINEAR REGRESSION FORMULA: Y = a + bX

weight = 0.7    # b
bias = 0.3    # a

start = 0
end = 1
step = 0.02

X = torch.arange(start, end, step).unsqueeze(dim=1)
Y = bias + (weight * X)
```

```python
plt.plot(X, Y)
plt.show()
```

```python
print(len(X), len(Y))
```

## Train-Test Split

```python
train_split = int(0.8 * len(X))    # 60%
test_split = int(0.2 * len(Y))   # 20% 20%

print(train_split, test_split)
```

```python
X_train = X[:train_split]
X_test = X[train_split:train_split + test_split]

print(len(X_train), len(X_test))
```

```python
Y_train = Y[:train_split]
Y_test = Y[train_split:train_split + test_split]

print(len(Y_train), len(Y_test))
```

```python
plt.scatter(X_train, Y_train, color='blue')
plt.scatter(X_test, Y_test, color='green')
plt.title('Train-Test-Validation Split')
```

## Creating our first PyTorch Model

```python
# CREATING A LINEAR REGRESSION MODEL CLASS

class LinearRegressionModel(nn.Module):    # Almost everything in PyTorch inherits from nn.Module
    def __init__(self):
        super().__init__()
        self.weights = nn.Parameter(torch.randn(1, requires_grad=True, dtype=torch.float))    # We can set requires grad to False when 
        self.bias = nn.Parameter(torch.randn(1, requires_grad=True, dtype=torch.float))       # we want to freeze certain layers (transfer learning).

    # Forward method to define computation in the model
    def forward(self, x: torch.Tensor) -> torch.Tensor:    # 'x' is the input data
        return self.weights * x + self.bias    # Linear regression formula
```

## Important model building classes

1) torch.nn:- Contains all the building blocks for computational graphs (another word for Neural Network).
2) torch.nn.Parameter:- What parameters should our model try to learn (often a PyTorch layer from torch.nn will set these for us).
3) torch.nn.Module:- The base class for all Neural Network modules.
4) def forward():- Defines what happens in each forward pass. (All nn.Module sub-classes require you to overwrite the forward() method).
5) torch.optim:- This is where the optimizers in PyTorch live. They will help with Gradient Descent.

```python
# WHAT'S INSIDE OUR MODEL?

torch.manual_seed(42)

lr_model = LinearRegressionModel()
print(list(lr_model.parameters()))
```

```python
# LIST NAMED PARAMETERS

lr_model.state_dict()
```

## Making prediction using torch.inference_mode()

```python
X_test
```

Inference mode: designed for inference scenarios where you want to make predictions or evaluate a model without needing to track gradients. 

```python
# CHECKING HOW WELL MODEL(UNTRAINED) PREDICTS y_test based on x_test

with torch.inference_mode():
    Y_pred = lr_model(X_test)

Y_pred
```
